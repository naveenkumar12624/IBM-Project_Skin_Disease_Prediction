```python
import os
import pandas as pd
import shutil
```

```python
for index, row in df.iterrows():
    image_name = row['isic_id'] + '.jpg'  # Assuming the image file extension is '.jpg'
    source_path = os.path.join(base_image_directory, image_name)
    destination_path = os.path.join(output_directory, row['diagnosis'], image_name)

    if pd.notna(row['diagnosis']):  # Check if 'diagnosis' is not null
        try:
            # Check if the source file exists before attempting to move it
            if os.path.isfile(source_path):
                # Move the image to the corresponding diagnosis folder
                shutil.move(source_path, destination_path)
            else:
                print(f"Source file not found: {source_path}. Skipping...")
        except Exception as e:
            print(f"Error occurred while moving file: {source_path}")
            print(f"Error details: {str(e)}")
    else:
        print(f"Null diagnosis value for file: {source_path}. Skipping...")
```
